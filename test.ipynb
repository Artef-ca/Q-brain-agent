```python
import json
from google.cloud import storage

def test_gcs_schema_fetch_fixed():
    print("🚀 Starting Fixed GCS Schema Test...")
    client = storage.Client()
    bucket = client.bucket("qbrain-staging")
    blobs = list(client.list_blobs(bucket, prefix="bq_schema_cards_"))
    
    for blob in blobs:
        if not blob.name.endswith(".jsonl"): continue
        content = blob.download_as_text()
        lines = [line for line in content.split('\n') if line.strip()]
        
        if lines:
            first_line = lines[0]
            
            # 1. First Parse (Outer Wrapper)
            outer_json = json.loads(first_line)
            
            # 2. Second Parse (Unpack jsonData)
            json_data_str = outer_json.get("jsonData", "{}")
            schema = json.loads(json_data_str)
            
            print("\n✅ SUCCESS! Unpacked jsonData keys:")
            print(list(schema.keys()))
            
            t_dataset = schema.get("dataset", "Unknown")
            t_table = schema.get("table", "Unknown")
            print(f"✅ Extracted Table Name: {t_dataset}.{t_table}")
            
            if "columns" in schema:
                print(f"✅ Extracted {len(schema['columns'])} columns.")
                # Let's check for samples in the second column (usually Cinema_ID in your data)
                if len(schema['columns']) > 1:
                    col2 = schema['columns'][1]
                    print(f"✅ Column '{col2.get('name')}' samples: {col2.get('sample_values')}")
            break

test_gcs_schema_fetch_fixed()
```

```python
import time
import statistics
import concurrent.futures
import vertexai
from vertexai.generative_models import GenerativeModel, GenerationConfig
from tqdm import tqdm # For progress bars

# --- CONFIGURATION ---
PROJECT_ID = "smartadvisor-483817"  # REPLACE WITH YOUR PROJECT ID

# Regions to test (Add/Remove as needed)
# "us-central1" (Iowa), "europe-west1" (Belgium), "me-central1" (Doha), etc.
REGIONS_TO_TEST = [
    "us-central1", 
    "us-east4",
    "europe-west1",
    "asia-southeast1", 
]

MODEL_NAME = "gemini-2.5-flash-001" # Flash is faster/cheaper for benchmarking
TOTAL_REQUESTS = 20    # Total calls per region (Increase to 50+ for heavy stress)
CONCURRENCY = 5        # How many requests to fire at EXACTLY the same time

PROMPT = "Explain the theory of relativity in 2 sentences."
# ---------------------

def make_request(model):
    """Performs a single LLM call and returns latency."""
    start = time.time()
    try:
        response = model.generate_content(
            PROMPT,
            generation_config=GenerationConfig(max_output_tokens=100)
        )
        # Access text to ensure request is fully complete
        _ = response.text 
        latency = time.time() - start
        return latency, None  # Success
    except Exception as e:
        return time.time() - start, str(e)  # Error

def test_region(region):
    """Stresses a specific region with concurrent requests."""
    print(f"\n--- Testing Region: {region} ---")
    
    # Initialize Vertex AI for this specific region
    try:
        vertexai.init(project=PROJECT_ID, location=region)
        model = GenerativeModel(MODEL_NAME)
    except Exception as e:
        print(f"Failed to init region {region}: {e}")
        return None

    latencies = []
    errors = []

    # Use ThreadPoolExecutor to simulate parallel users
    with concurrent.futures.ThreadPoolExecutor(max_workers=CONCURRENCY) as executor:
        # Submit all tasks
        futures = [executor.submit(make_request, model) for _ in range(TOTAL_REQUESTS)]
        
        # Process as they complete
        for future in tqdm(concurrent.futures.as_completed(futures), total=TOTAL_REQUESTS, desc=f"Stressing {region}"):
            latency, error = future.result()
            if error:
                errors.append(error)
            else:
                latencies.append(latency)

    return {
        "region": region,
        "avg_latency": statistics.mean(latencies) if latencies else 0,
        "p95_latency": statistics.quantiles(latencies, n=20)[18] if len(latencies) >= 20 else 0,
        "success_count": len(latencies),
        "error_count": len(errors),
        "errors": errors[:3] # Store first 3 errors for debugging
    }

def main():
    results = []
    
    print(f"Starting Benchmark: {TOTAL_REQUESTS} reqs/region | {CONCURRENCY} concurrent workers")
    print(f"Model: {MODEL_NAME}\n")

    for region in REGIONS_TO_TEST:
        data = test_region(region)
        if data:
            results.append(data)
            # Cool down slightly between regions to avoid global quota side-effects
            time.sleep(2) 

    # --- PRINT RESULTS TABLE ---
    print("\n" + "="*85)
    print(f"{'REGION':<20} | {'SUCCESS':<8} | {'ERRORS':<8} | {'AVG LATENCY (s)':<15} | {'P95 LATENCY (s)':<15}")
    print("="*85)
    
    # Sort by Fastest Average Latency
    results.sort(key=lambda x: x['avg_latency'])

    for r in results:
        print(f"{r['region']:<20} | {r['success_count']:<8} | {r['error_count']:<8} | {r['avg_latency']:.4f}          | {r['p95_latency']:.4f}")
    print("="*85)

    if results and results[0]['error_count'] == 0:
        print(f"\n🏆 Winner: {results[0]['region']} is the fastest stable region for you.")
    elif results:
         print(f"\n⚠️ Note: {results[0]['region']} was fastest, but check error rates.")

if __name__ == "__main__":
    main()
```

```python
from pathlib import Path

def generate_tree(dir_path: Path, prefix: str = '', ignore_dirs: list = None):
    if ignore_dirs is None:
        ignore_dirs = ['.git', '.ipynb_checkpoints', '__pycache__', 'node_modules']
        
    # Get all items, filter out ignored directories and hidden files, then sort
    items = sorted([
        p for p in dir_path.iterdir() 
        if p.name not in ignore_dirs and not p.name.startswith('.')
    ], key=lambda x: x.name.lower())
    
    pointers = ['├── '] * (len(items) - 1) + ['└── '] if items else []
    
    for pointer, item in zip(pointers, items):
        print(prefix + pointer + item.name)
        if item.is_dir():
            # Extend the prefix for the next level down
            extension = '│   ' if pointer == '├── ' else '    '
            generate_tree(item, prefix=prefix + extension, ignore_dirs=ignore_dirs)

# Usage: Set the path to your current directory or target folder
target_folder = Path('.') 
print(target_folder.name)
generate_tree(target_folder)
```
