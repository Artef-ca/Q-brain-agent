```python


from google.cloud import bigquery

# ─────────────────────────────────────────────
# CONFIG — update these values
# ─────────────────────────────────────────────
PROJECT_ID   = "prj-ai-dev-qic"
DATASET_ID   = "qic_hr"
TABLE_ID     = "dim_department"
client = bigquery.Client(project=PROJECT_ID)
# New table-level description (set to None to leave unchanged)
NEW_TABLE_DESCRIPTION = "This table stores hr data"

# Map of column_name → new description.
# Only listed columns will be updated; all others are left as-is.
COLUMN_DESCRIPTIONS = {
    "Department_ID":   "Department ID",
    "Department_Name":   "Department Name",
    "Parent_Department_ID": "Parent Department ID",
}
# ─────────────────────────────────────────────


def update_descriptions(
    project: str,
    dataset: str,
    table: str,
    table_desc: str | None,
    column_descs: dict[str, str],
) -> None:
    
    table_ref = f"{project}.{dataset}.{table}"

    # 1. Fetch the existing table object (full schema preserved)
    bq_table = client.get_table(table_ref)
    print(f"✅ Fetched table: {table_ref}")
    print(f"   Rows: {bq_table.num_rows:,}  |  Columns: {len(bq_table.schema)}")

    # 2. Rebuild schema — only touch descriptions, nothing else
    new_schema = []
    for field in bq_table.schema:
        if field.name in column_descs:
            # Reconstruct the field with the new description, keeping everything else
            updated_field = bigquery.SchemaField(
                name        = field.name,
                field_type  = field.field_type,
                mode        = field.mode,
                description = column_descs[field.name],
                fields      = field.fields,           # preserves nested/repeated fields
                precision   = field.precision,
                scale       = field.scale,
                max_length  = field.max_length,
            )
            new_schema.append(updated_field)
            print(f"   📝 Updated column: {field.name}")
        else:
            new_schema.append(field)                  # no change

    # 3. Apply schema update
    bq_table.schema = new_schema

    # 4. Update table description if provided
    fields_to_update = ["schema"]
    if table_desc is not None:
        bq_table.description = table_desc
        fields_to_update.append("description")
        print(f"   📝 Updated table description")

    # 5. Patch the table (only schema + description fields are touched)
    client.update_table(bq_table, fields_to_update)
    print(f"\n✅ Done — table updated successfully: {table_ref}")


update_descriptions(
    project     = PROJECT_ID,
    dataset     = DATASET_ID,
    table       = TABLE_ID,
    table_desc  = NEW_TABLE_DESCRIPTION,
    column_descs= COLUMN_DESCRIPTIONS,
)
```
